```python
#1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".
class Animal:
    def speak(self):
        print("This is a generic animal sound.")

class Dog(Animal):
    def speak(self):
        print("Bark!")

a1 = Animal()
a1.speak()

b1 = Dog()
b1.speak()

#2
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.a1 = radius

    def area(self):
        return 3.14 * self.a1 * self.a1

class Rectangle(Shape):
    def __init__(self, length, width):
        self.a2 = length
        self.b2 = width

    def area(self):
        return self.a2 * self.b2

c1 = Circle(5)
print(c1.area())

r1 = Rectangle(4, 6)
print(r1.area())

#3
class Vehicle:
    def __init__(self, type):
        self.a3 = type

class Car(Vehicle):
    def __init__(self, type, model):
        super().__init__(type)
        self.b3 = model

class ElectricCar(Car):
    def __init__(self, type, model, battery):
        super().__init__(type, model)
        self.c3 = battery

v1 = ElectricCar("Electric", "Tesla", "100kWh")
print(v1.a3, v1.b3, v1.c3)

#5
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance.")

    def check_balance(self):
        return self.__balance

a4 = BankAccount()
a4.deposit(1000)
a4.withdraw(500)
print(a4.check_balance())

#6
class Instrument:
    def play(self):
        print("Playing an instrument.")

class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar.")

class Piano(Instrument):
    def play(self):
        print("Playing the piano.")

a5 = Instrument()
b5 = Guitar()
c5 = Piano()

a5.play()
b5.play()
c5.play()

#7
class MathOperations:
    @classmethod
    def add_numbers(cls, a6, b6):
        return a6 + b6

    @staticmethod
    def subtract_numbers(a7, b7):
        return a7 - b7

print(MathOperations.add_numbers(10, 5))
print(MathOperations.subtract_numbers(10, 5))

#8
class Person:
    count = 0

    def __init__(self):
        Person.count += 1

    @classmethod
    def total_persons(cls):
        return cls.count

a8 = Person()
b8 = Person()
print(Person.total_persons())

#9
class Fraction:
    def __init__(self, numerator, denominator):
        self.a9 = numerator
        self.b9 = denominator

    def __str__(self):
        return f"{self.a9}/{self.b9}"

f1 = Fraction(3, 4)
print(f1)

#10
class Vector:
    def __init__(self, x, y):
        self.a10 = x
        self.b10 = y

    def __add__(self, other):
        return Vector(self.a10 + other.a10, self.b10 + other.b10)

    def __str__(self):
        return f"({self.a10}, {self.b10})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3)

#11
class Person:
    def __init__(self, name, age):
        self.a11 = name
        self.b11 = age

    def greet(self):
        print(f"Hello, my name is {self.a11} and I am {self.b11} years old.")

p1 = Person("Alice", 25)
p1.greet()

#12
class Student:
    def __init__(self, name, grades):
        self.a12 = name
        self.b12 = grades

    def average_grade(self):
        return sum(self.b12) / len(self.b12)

s1 = Student("Bob", [85, 90, 78, 92])
print(s1.average_grade())

#13
class Rectangle:
    def set_dimensions(self, length, width):
        self.a13 = length
        self.b13 = width

    def area(self):
        return self.a13 * self.b13

r1 = Rectangle()
r1.set_dimensions(5, 10)
print(r1.area())

#14
class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.a14 = hours_worked
        self.b14 = hourly_rate

    def calculate_salary(self):
        return self.a14 * self.b14

class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.c14 = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.c14

e1 = Employee(40, 20)
print(e1.calculate_salary())

m1 = Manager(40, 20, 500)
print(m1.calculate_salary())

#15
class Product:
    def __init__(self, name, price, quantity):
        self.a15 = name
        self.b15 = price
        self.c15 = quantity

    def total_price(self):
        return self.b15 * self.c15

p1 = Product("Laptop", 1000, 2)
print(p1.total_price())

#16
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo!")

class Sheep(Animal):
    def sound(self):
        print("Baa!")

c1 = Cow()
s1 = Sheep()

c1.sound()
s1.sound()

#17
class Book:
    def __init__(self, title, author, year_published):
        self.a17 = title
        self.b17 = author
        self.c17 = year_published

    def get_book_info(self):
        return f"Title: {self.a17}, Author: {self.b17}, Year Published: {self.c17}"

b1 = Book("1984", "George Orwell", 1949)
print(b1.get_book_info())

#18
class House:
    def __init__(self, address, price):
        self.a18 = address
        self.b18 = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.c18 = number_of_rooms

h1 = House("123 Main St", 300000)
m1 = Mansion("456 Elm St", 1000000, 10)

print(h1.a18, h1.b18)
print(m1.a18, m1.b18, m1.c18)
```
